Goal is to analyze RNA-seq data to determine whether ectopic expression of Xist-872-Airn from Rosa26 impacts gene expression (relative to RMCE-empty (dox treated), also to no-dox Xist872-Airn cells).

For reference, using Jupyter notebook 2023-04-11_RMCE-Airn_Xist_RNA-seq

```python
#list what is in my working directory
os.listdir()
```

```python
import pandas as pd
#want to be able to see all columns in large dataframes
pd.set_option('display.max_columns', None)

#read in my large data file containing B6/CAST-parsed reads (exon, ejc, intron, total) for each sample
#This text file was copied into my computer's working directory Bioinformatics\\2023-02-06_RMCE-Airn_RNA-seq, shown above
#pd.read_table reads in tab-delimited text files by default
df = pd.read_table('2024-04-30_RMCE_Xist872-Airn_RNA-seq_SNP_reads_all_data.txt')
df
```

```python
#For my first analysis, I want to subset this dataframe down to only the "B6_total" columns
#make dataframe containing ther first 6 "identifier" columns - specify start (inclusive) of 0, end (NOT INCLUSIVE!) of 6
df_ID_columns = df.iloc[:,0:6]
df_ID_columns
```

```python
#make dataframe containing only the "B6_total" columns
df_B6_total_data = df.loc[:, df.columns.str.contains('B6_total')]
df_B6_total_data
```

```python
#Now combine df_ID_columns and df_B6_total_data

df_B6_total = pd.concat([df_ID_columns, df_B6_total_data], axis=1)
df_B6_total
```

I want to run DESeq2 (R package) to identify differentially expressed genes between the following groups of samples:

1. RMCE-empty (4 clones, 1 sample each) vs. RMCE-Xist872-Airn 1000 ng/mL dox (2 clones, 2 samples each)
2. RMCE-Xist872-Airn 1000 ng/mL dox (2 clones, 2 samples each) vs. RMCE-Xist872-Airn no dox (2 clones, 2 samples each)

```python
df_B6_total_empty_data = df_B6_total.loc[:, df_B6_total.columns.str.contains('empty')]
df_B6_total_XA_dox_data = df_B6_total[['XA1_dox_rep1_RNAseq_B6_total', 'XA1_dox_rep2_RNAseq_B6_total', 'XA2_dox_rep1_RNAseq_B6_total', 'XA2_dox_rep2_RNAseq_B6_total',]]

df_B6_total_empty_XA_dox = pd.concat([df_ID_columns, df_B6_total_empty_data, df_B6_total_XA_dox_data,], axis=1)
df_B6_total_empty_XA_dox
```

Last step before running DESeq2: pre-filter the genes to include only those that have an average of 5 reads per sample (so a total of 40 reads across these 8 samples).

```python
df_B6_total_empty_XA_dox_filtered = df_B6_total_empty_XA_dox[df_B6_total_empty_XA_dox.loc[:, "empty1_RNAseq_B6_total" : "XA2_dox_rep2_RNAseq_B6_total"].sum(axis=1) >= 40]
df_B6_total_empty_XA_dox_filtered
```

For DESeq2, will need this sata in the form of a matrix. Matrices in R can have row names and column names - will want a single rowname, so I will use the "id" column here:

```python
df_B6_total_empty_XA_dox_filtered2 = df_B6_total_empty_XA_dox_filtered.set_index('id')
df_B6_total_empty_XA_dox_filtered2
```

```python
df_B6_total_empty_XA_dox_filtered2.index.names = [None]
df_B6_total_empty_XA_dox_filtered2
```

```python
#Now remove the ID/metadata columns
df_B6_total_empty_XA_dox_filtered3 = df_B6_total_empty_XA_dox_filtered2.iloc[:, 5:14]
df_B6_total_empty_XA_dox_filtered3
```

```python
df_B6_total_empty_XA_dox_filtered3.to_csv('2024-04-30_RMCE_B6_reads_empty_XA_dox_for_DESeq2.csv')
```

DESeq2 (in R) shows that of 728 expressed chr6 genes, 9 are significantly up and 63 are significantly down!! For comparison, we previously found the following numbers of genes downregulated (out of those calculated as "expressed for each comparison vs. empty): Airn (0/710), Xist 10 ng/mL dox (21/706), Xist 1000 ng/mL dox (242/671).
